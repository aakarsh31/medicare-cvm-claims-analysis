## Step 0. Prepare raw input datasets

Here we will 1) load original datasets, 2) remove unused columns, 3) de-duplicate rows, and 4) join datasets, not necessarily in this order

Assumptions:
- Claim ID (cur_clm_uniq_id) represents one claim, which may or may not have more than one code (code could be HCPCS/CPT, diagnosis, procedure...)
- There is a one-to-many relationship between patient IDs (bene_mbi_id) and claim IDs (cur_clm_uniq_id), i.e. each claim is unique to one patient, but one patient can have more than one claim

### 0.3 Combine all datasets

### 0.3.1. Join datasets with common records

```python
# Join Claims Header and Claims Revenue Center datasets on claim ID, patient ID, and claim date
# Perform outer join to capture all possible claims
medicare_df = pd.merge(
    parta_claims_header_df,
    parta_claims_revenue_center_detail_df,
    on=['claim_id','patient_id','claim_date'], how='outer'
)
medicare_df.sort_values(by='claim_id')#.head(100)
```

```python
# Since Claims Header dataset has some principal diagnosis codes and the Diagnosis dataset
#  supplements them with additional codes wherever possible,
#  coalesce them with preference to the principal code from Claim Header dataset
# Once the diagnosis codes are combined into one column, remove the older columns and any duplicates
medicare_df['diagnosis_code'] = medicare_df[['prncpl_dgns_cd', 'clm_dgns_cd']].bfill(axis=1).iloc[:, 0]
medicare_df = medicare_df.drop(['prncpl_dgns_cd', 'clm_dgns_cd'], axis=1).drop_duplicates()
medicare_df.sort_values(by='claim_id').head(20)
```

```python
#Applying an inner join to take only codes in top100 into consideration
inner_merged_top100_df = pd.merge(medicare_df,hcpcs_code_100_df, on = 'hcpcs_code',how = 'inner')

#Create 'is_CVM' column differentiating cardiac from non cardiac codes
inner_merged_top100_df['is_CVM'] = np.where(inner_merged_top100_df['category']=='Cardiac',1,0)
inner_merged_top100_df.head()
```

```python
# Collapsing claim lines to the claim level, if any HCPCS Claim has Cardiac
# The entire claim is classified as is_CVMS, done by using max() to ensure
# a claim with both cvm and non-cvm procedures is counted once.

claim_level_df = inner_merged_top100_df.groupby(['claim_year', 'claim_id']).agg(
    is_CVM=('is_CVM', 'max')
).reset_index()

claim_level_df
```

```python
#Counts per year per CVM/non-CVM claims
cvm_counts = claim_level_df.groupby(['claim_year', 'is_CVM']).agg(clm_cnt=('claim_id', 'nunique')).reset_index()
cvm_counts
```

```python
#Total claims per year
total_per_year = claim_level_df.groupby('claim_year').agg(total_clm_cnt=('claim_id', 'nunique')
).reset_index()
total_per_year
```

```python
cvm_percnt = pd.merge(cvm_counts, total_per_year, on='claim_year')
cvm_percnt['percentage'] = cvm_percnt['clm_cnt']/cvm_percnt['total_clm_cnt']*100
cvm_percnt
```

```python
import matplotlib.pyplot as plt

cvm_pivot = cvm_percnt.pivot(index='claim_year', columns='is_CVM', values='percentage')

fig, ax = plt.subplots(figsize=(8, 6))
ax.bar(cvm_pivot.index, cvm_pivot[0], label='Non-CVM', color='#AEC6CF')
ax.bar(cvm_pivot.index, cvm_pivot[1], bottom=cvm_pivot[0], label='CVM', color='#F4A6A0')
ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('Percentage of Claims (%)', fontsize=12)
ax.set_title('Share of CVM vs Non-CVM Claims (2016-2018)', fontsize=14, fontweight='bold')
ax.legend(fontsize=11, loc='upper center', bbox_to_anchor=(0.5, -0.1), ncol=2)
ax.set_ylim(0, 100)
ax.set_xticks(cvm_pivot.index)

for year in cvm_pivot.index:
    non_cvm_val = cvm_pivot.loc[year, 0]
    cvm_val = cvm_pivot.loc[year, 1]
    ax.text(year, non_cvm_val / 2, f'{non_cvm_val:.1f}%', ha='center', va='center', fontsize=11, fontweight='bold')
    ax.text(year, non_cvm_val + cvm_val / 2, f'{cvm_val:.1f}%', ha='center', va='center', fontsize=11, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
#Function to assign age segments
def assign_age_segment(age):
    if age <18:
        return None
    elif age <= 59:
        return '18-59'
    elif age <= 69:
        return '60-69'
    elif age <= 79:
        return '70-79'
    else:
        return '80+'

inner_merged_top100_df_cvm_cleaned['age_segment'] = inner_merged_top100_df_cvm_cleaned['patient_age'].apply(assign_age_segment)
inner_merged_top100_df_cvm_cleaned
```

```python
#Removing any instances of ages under 18
inner_merged_top100_df_cvm_cleaned = inner_merged_top100_df_cvm_cleaned.dropna(subset=['age_segment'])
```

```python
#Group by age segments per year
age_segment_counts = inner_merged_top100_df_cvm_cleaned.groupby(['claim_year','age_segment']).agg(clm_cnt=('claim_id', 'nunique')).reset_index()
age_segment_counts
```

```python
#Stacked bar Graph for age segment distribution
age_pivot = age_segment_counts.pivot(index='claim_year', columns='age_segment', values='clm_cnt')
age_pivot = age_pivot[['18-59', '60-69', '70-79', '80+']]
years = age_pivot.index.astype(int)
age_pivot.index = years

#Calculate YoY change percentage
yoy_pct = age_pivot.pct_change() * 100
yoy_pct = yoy_pct.dropna()  # Removing 2016 since no prior year

fig, (ax, ax_table1, ax_table2) = plt.subplots(3, 1, figsize=(10, 11), gridspec_kw={'height_ratios': [3, 1, 1]})

bottom = np.zeros(len(age_pivot))
colors = ['#AEC6CF', '#F4A6A0', '#B5EAD7', '#FFDAC1']
segments = age_pivot.columns

for i, seg in enumerate(segments):
    vals = age_pivot[seg].values
    ax.bar(age_pivot.index, vals, bottom=bottom, label=seg, color=colors[i])

    for j, year in enumerate(age_pivot.index):
        val = vals[j]
        mid = bottom[j] + val / 2
        ax.text(year, mid, str(int(val)), ha='center', va='center', fontsize=10, fontweight='bold')

    bottom += vals

ax.set_xlabel('Year', fontsize=12)
ax.set_ylabel('Number of Claims', fontsize=12)
ax.set_title('CVM Claims by Patient Age Group (2016-2018)', fontsize=14, fontweight='bold')
ax.legend(fontsize=10, loc='upper center', bbox_to_anchor=(0.5, -0.06), ncol=4)
ax.set_xticks(years)

#Raw Counts Table
ax_table1.axis('off')
ax_table1.set_title('Claim Counts by Age Group', fontsize=11, fontweight='bold', pad=10)
table1 = ax_table1.table(
    cellText=age_pivot.values.astype(int).tolist(),
    rowLabels=[str(y) for y in years],
    colLabels=list(segments),
    cellLoc='center',
    loc='center',
    colColours=colors
)
table1.auto_set_font_size(False)
table1.set_fontsize(10)
table1.scale(1, 1.5)

#YoY change percentage Table
ax_table2.axis('off')
ax_table2.set_title('Year-over-Year Change (%)', fontsize=11, fontweight='bold', pad=10)
yoy_data = [[f'{val:.1f}%' for val in row] for row in yoy_pct.values]
table2 = ax_table2.table(
    cellText=yoy_data,
    rowLabels=[str(y) for y in yoy_pct.index],
    colLabels=list(segments),
    cellLoc='center',
    loc='center',
    colColours=colors
)
table2.auto_set_font_size(False)
table2.set_fontsize(10)
table2.scale(1, 1.5)

plt.tight_layout()
plt.show()
```
